```python
from stats_can import StatsCan

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#seaborn theme
import seaborn as sns
sns.set()
```

```python
sc = StatsCan()
```

```python
sc
```

```python
df = sc.table_to_df("17-10-0136-01")
```

```python
df
```

```python
calgary = df[df['GEO']=="Calgary (CMA), Alberta"]
```

```python
calgary
```

```python
calgary['Components of population growth'].unique()
```

```python
calgary['STATUS'].unique()
```

```python
calgary = calgary.filter(items = ['REF_DATE','Components of population growth','Sex','Age group', 'VALUE'])
```

```python
calgary
```

```python
calgary[(calgary['Sex']=='Both sexes') & (calgary['Age group'] == 'All ages')].pivot(index='REF_DATE', columns = 'Components of population growth',values ='VALUE')
```

```python
cal_frame = calgary[(calgary['Sex']=='Both sexes') & (calgary['Age group'] == 'All ages')].pivot(index='REF_DATE', columns = 'Components of population growth',values ='VALUE')
```

```python
cal_frame
```

```python
cal_frame.index
```

```python

cal_frame.plot(figsize=(12,6),kind='bar')
```

```python
cal_frame.plot(figsize=(16,6),kind='bar')
```

```python
plt.figure(figsize=(20,14))
sns.barplot(x='REF_DATE',y = 'VALUE', hue='Components of population growth', data= calgary[(calgary['Sex']=='Both sexes') & (calgary['Age group'] == 'All ages')])
```

```python
cal_frame['Net_growth']=cal_frame['Births']-cal_frame['Deaths']
cal_frame['Net_immigration']=cal_frame['Immigrants']-cal_frame['Emigrants']
cal_frame['Net_pop']=cal_frame['Net interprovincial migration']+cal_frame['Net intraprovincial migration'] + cal_frame['Net_immigration'] + cal_frame['Net non-permanent residents'] + cal_frame['Returning emigrants']
```

```python
cal_frame[['Net interprovincial migration','Net_growth', 'Net non-permanent residents', 'Net_immigration', 'Net intraprovincial migration', 'Immigrants']].plot(figsize=(20,14),kind='bar')
```

```python
cal_frame
```

```python

plt.figure(figsize=(20,14))
sns.lineplot(x=cal_frame.index, y = 'Net_pop', data = cal_frame)
```

```python
cal_frame.fillna(value=0, inplace=True)
cal_frame['tot_migrate']=cal_frame['Net_pop']-cal_frame['Net temporary emigration']+cal_frame['Residual deviation']
```

```python
cal_frame
```

```python

plt.figure(figsize=(20,14))
sns.lineplot(x=cal_frame.index, y = 'Net_pop', data = cal_frame)
sns.lineplot(x=cal_frame.index, y = 'tot_migrate', data = cal_frame)
```

```python
cal_frame['Net_pop'].to_csv('total_migration.csv')
```
