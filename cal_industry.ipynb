```python
from stats_can import StatsCan

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#seaborn theme
import seaborn as sns
sns.set()
```

```python
sc = StatsCan()
```

```python
df= sc.table_to_df('14-10-0098-01')
```

```python
df
```

```python
calgary= df[df['GEO']=='Calgary, Alberta']
```

```python
calgary['UOM'].unique()
```

```python
calgary
```

```python
cal_data = calgary.filter(items = ['REF_DATE','North American Industry Classification System (NAICS)','SCALAR_FACTOR','UOM', 'VALUE'])
```

```python
cal_data['SCALAR_FACTOR'].unique()
```

```python
cal_pivot = cal_data.pivot(index='REF_DATE', columns = 'North American Industry Classification System (NAICS)',values ='VALUE')
```

```python
cal_pivot
```

```python
cal_pivot[['Forestry, fishing, mining, quarrying, oil and gas [21, 113-114, 1153, 2100]','Services-producing sector','Goods-producing sector']].plot(figsize=(20,15))
```
