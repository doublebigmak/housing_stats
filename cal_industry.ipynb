```python
from stats_can import StatsCan

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#seaborn theme
import seaborn as sns
sns.set(font='Gill Sans MT')
```

```python
sc = StatsCan()
```

```python
sc.update_tables()

```

```python
df= sc.table_to_df('14-10-0384-01')
```

```python
df
```

```python
calgary= df[df['GEO']=='Calgary, Alberta']
```

```python
calgary['UOM'].unique()
```

```python
calgary
```

```python
cal_data = calgary.filter(items = ['REF_DATE','North American Industry Classification System (NAICS)','SCALAR_FACTOR','UOM', 'VALUE'])
```

```python
cal_data['SCALAR_FACTOR'].unique()
```

```python
cal_pivot = cal_data.pivot(index='REF_DATE', columns = 'North American Industry Classification System (NAICS)',values ='VALUE')
```

```python
cal_pivot
```

```python
cal_pivot[['Forestry, fishing, mining, quarrying, oil and gas [21, 113-114, 1153, 2100]','Services-producing sector','Goods-producing sector']].plot(figsize=(20,15))
```

```python
df_3ma = sc.table_to_df('14-10-0379-01')
```

```python
df_3ma
```

```python
calgary_3ma = df_3ma[df_3ma['GEO']=='Calgary, Alberta']
```

```python
cal_data_3ma = calgary_3ma.filter(items = ['REF_DATE','North American Industry Classification System (NAICS)','SCALAR_FACTOR','UOM', 'VALUE'])
cal_pivot_3ma = cal_data_3ma.pivot(index='REF_DATE', columns = 'North American Industry Classification System (NAICS)',values ='VALUE')
```

```python
cal_pivot_3ma['2019':]
```

```python
cal_pivot_3ma[['Goods-producing sector','Services-producing sector']].plot(figsize=(16,9),xlabel='Date')
plt.tight_layout()
plt.savefig('goods_service.png')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2,figsize=(20,15))
cal_pivot_3ma[['Forestry, fishing, mining, quarrying, oil and gas [21, 113-114, 1153, 2100]','Manufacturing [31-33]','Professional, scientific and technical services [54]',"Wholesale and retail trade [41-45]",'Health care and social assistance [62]','Accommodation and food services [72]','Construction [23]','Public administration [91]']].plot(ax= axes,subplots=True)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2,figsize=(20,15))
cal_pivot_3ma['2019':][['Forestry, fishing, mining, quarrying, oil and gas [21, 113-114, 1153, 2100]','Manufacturing [31-33]','Professional, scientific and technical services [54]',"Wholesale and retail trade [41-45]",'Health care and social assistance [62]','Accommodation and food services [72]','Construction [23]']].plot(ax=axes, subplots=True)
```

```python
cal_pivot_3ma[['Forestry, fishing, mining, quarrying, oil and gas [21, 113-114, 1153, 2100]','Manufacturing [31-33]','Professional, scientific and technical services [54]',"Wholesale and retail trade [41-45]",'Health care and social assistance [62]', 'Accommodation and food services [72]']].plot.area(stacked = True,figsize=(20,15))
```

```python
cal_pivot_3ma[['Goods-producing sector','Services-producing sector']].plot.area(stacked=True,figsize=(16,9),xlabel='Date')
plt.tight_layout()
plt.savefig('goods_service_stack.png')
```
