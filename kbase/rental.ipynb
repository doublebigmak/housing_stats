General Rental data - Calgary CMA

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font='Arial')
```

```python
vacancy = pd.read_csv('..\\data\\pareto\\vac_zerobd.csv',index_col=0)
vacancy=vacancy.join(pd.read_csv('..\\data\\pareto\\vac_onebd.csv',index_col=0))
vacancy=vacancy.join(pd.read_csv('..\\data\\pareto\\vac_twobd.csv',index_col=0))
vacancy=vacancy.join(pd.read_csv('..\\data\\pareto\\vac_thrbd.csv',index_col=0))
vacancy=vacancy.join(pd.read_csv('..\\data\\pareto\\vac_tot.csv',index_col=0))
vacancy.index=vacancy.index.rename('Date')

vacancy=vacancy.dropna()
```

## Vacancy Rates

```python
vacancy['vac_tot'].plot(title='Total Vacancy Rate, Calgary CMA', figsize=(16,10))
plt.legend(['Total Vacancy Rate'])
plt.tight_layout()
plt.savefig('kbase_graphs/vac_tot.png')
```

```python
vacancy[['vac_zerobd','vac_onebd','vac_twobd','vac_thrbd']].plot(title='Vacancy Rates by Bedroom Type, Calgary CMA', figsize=(16,10))
plt.legend(['Bachelors','One Bed','Two Bed','Three Bed'])
plt.tight_layout()
plt.savefig('kbase_graphs/vac_bed.png')
```

Vacancy rates by bedroom types - note that three bed+ units only regularly had the highest vacancy rates starting in 2005. Before that, bachelor units had the highest vacancy rates across all types. It is important to note however, the majority of the rental universe is in one and two bed units. 


## Rental Rates

```python
rents = pd.read_csv('..\\data\\pareto\\rent_zerobd.csv',index_col=0)
rents=rents.join(pd.read_csv('..\\data\\pareto\\rent_onebd.csv',index_col=0))
rents=rents.join(pd.read_csv('..\\data\\pareto\\rent_twobd.csv',index_col=0))
rents=rents.join(pd.read_csv('..\\data\\pareto\\rent_thrbd.csv',index_col=0))
rents=rents.join(pd.read_csv('..\\data\\pareto\\rent_tot.csv',index_col=0))
rents.index=rents.index.rename('Date')

rents=rents.dropna()
```

```python
rents['rent_tot'].plot(title='Average Rental Rate, Calgary CMA', figsize=(16,10))
plt.legend(['Average Rental Rate'])
plt.tight_layout()
plt.savefig('kbase_graphs/rent_tot.png')
```

Averge rental rates reflects both increasing home prices, inflation, and the boom-bust cycle. While there is some decreases in average rent during regimes of high vacancy rates, the overall trend is still higher.

```python
rents[['rent_zerobd','rent_onebd','rent_twobd','rent_thrbd']].plot(title='Average Rental Rates by Bedroom Type, Calgary CMA', figsize=(16,10))
plt.legend(['Bachelors','One Bed','Two Bed','Three Bed'])
plt.tight_layout()
plt.savefig('kbase_graphs/rent_bed.png')
```

Rental rates across bedrooms, and changes show demand and availability of those units. We can see that while one bed rents haven't reached 2014 levels, 3 bed and 2 bed average rents have surpassed that level in 2021, mainly due to newer units coming on to market.

## Rental Universe

```python
universe = pd.read_csv('..\\data\\pareto\\uni_zerobd.csv',index_col=0)
universe=universe.join(pd.read_csv('..\\data\\pareto\\uni_onebd.csv',index_col=0))
universe=universe.join(pd.read_csv('..\\data\\pareto\\uni_twobd.csv',index_col=0))
universe=universe.join(pd.read_csv('..\\data\\pareto\\uni_thrbd.csv',index_col=0))
universe=universe.join(pd.read_csv('..\\data\\pareto\\uni_tot.csv',index_col=0))
universe.index=universe.index.rename('Date')

universe=universe.dropna()
```

```python
universe['uni_tot'].plot(title='Rental Universe, Calgary CMA', figsize=(16,10))
plt.legend(['Total Rental Universe'])
plt.tight_layout()
plt.savefig('kbase_graphs/uni_tot.png')
```

Calgary's rental universe was on a decreasing trend since data was available in the early 1990s. This trend didn't reverse until the early 2010s, in 2013. The increase in rental universe is mainly due to newer developments coming online, as developers catchup to rental demand from the oil boom of the 2000s, and later early 2010s. 

```python
universe[['uni_zerobd','uni_onebd','uni_twobd','uni_thrbd']].plot(title='Rental Universe by Bedroom Type, Calgary CMA', figsize=(16,10))
plt.legend(['Bachelors','One Bed','Two Bed','Three Bed'])
plt.tight_layout()
plt.savefig('kbase_graphs/uni_bed.png')
```

As you can see, the majority of rental stock in Calgary exists as one and two bedroom type units. Two bedroom type units can be seen growing at a faster pace than one-bed room types in recent years, possibly reflecting demand preference, either to live in the inner-city, or to have more space.
