```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
rental = gpd.read_file("./geo/\CMHC_2020_shp/ZONE_2020.shp")
```

```python
calgary = rental[rental['MET_CODE']=='0140']
```

```python
calgary
```

```python
rental_data = pd.read_clipboard()
```

```python
calgary=calgary.merge(rental_data, left_on='ZONE_NAME_',right_on='Zone')
```

```python
fig,ax = plt.subplots(1,1,figsize=(20,15))
plt.rc('font',size=24)
ax.axis('off')
plot = calgary.plot(column='Change in Vacancy Rate',cmap='RdYlGn_r',edgecolor='black', ax=ax,alpha = 0.4, legend = True)
for x, y, label in zip(calgary.geometry.centroid.x, calgary.geometry.centroid.y, calgary['Change in Vacancy Rate']):
    ax.annotate("{:,.1f}".format(label), xy=(x, y), xytext=(1, 1), textcoords="offset points",fontsize = 24)

plt.tight_layout()
plt.savefig('rental_change_2022.png',dpi=300)

```
