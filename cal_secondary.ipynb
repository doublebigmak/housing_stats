```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd

#seaborn theme
import seaborn as sns
sns.set()
sns.set(font='Gill Sans MT')
```

https://data.calgary.ca/Business-and-Economic-Activity/Secondary-Suites/jwn6-r58y for data

```python
s_suite = pd.read_csv('data//Secondary_Suites.csv')
```

```python
s_suite['InspectionDate']=pd.to_datetime(s_suite['InspectionDate'])
s_suite['CompletionDate']=pd.to_datetime(s_suite['CompletionDate'])
s_suite=s_suite.set_index('CompletionDate')
```

```python
s_suite
```

```python
s_suite['2017':'2020'].resample('M').count()['id'].plot(title='Secondary Suites Approved', figsize = (14,8))
plt.savefig('secondary_suites_approve.png')
```

```python
st_rental = pd.read_csv('data//Short_Term_Rentals.csv')
```

https://data.calgary.ca/Business-and-Economic-Activity/Short-Term-Rentals/gzkz-5k9a for data

```python
st_rental
```

```python
st_rental['Licenced Date']=pd.to_datetime(st_rental['Licenced Date'])
st_rental['Licenced Expiry Date']=pd.to_datetime(st_rental['Licenced Expiry Date'])
st_rental=st_rental.set_index('Licenced Date')
```

```python
st_rental.resample('Q').count()['Business ID'].cumsum().plot(title='Cumulative Short-Term Rental Approved', figsize = (10,8), kind='bar')
plt.tight_layout()
plt.savefig('short_term_rental.png')
```
