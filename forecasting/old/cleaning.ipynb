```python
cal_high = pd.read_csv('https://open.alberta.ca/dataset/5c856875-786c-40f8-822e-a2d4bd305d68/resource/49dcf9f5-7814-478e-a7ea-df41d3639814/download/2020-2046-16-economic-region-population-projections-high.csv')
cal_med = pd.read_csv('https://open.alberta.ca/dataset/5c856875-786c-40f8-822e-a2d4bd305d68/resource/330b658b-ca2c-49c1-8114-16df919d13ff/download/2020-2046-14-economic-region-population-projections-med.csv')
cal_low = pd.read_csv('https://open.alberta.ca/dataset/5c856875-786c-40f8-822e-a2d4bd305d68/resource/6cb4bd3b-3f29-413b-919b-8840eae97ad0/download/2020-2046-12-economic-region-population-projections-low.csv')

cal_high = cal_high[cal_high['Region']=='6']
cal_med = cal_med[cal_med['Region']=='6']
cal_low = cal_low[cal_low['Region']=='6']

cal_high.index = cal_high['Year']
cal_med.index = cal_med['Year']
cal_low.index = cal_low['Year']

cal_high_all=cal_high[((cal_high['Age']=='Total') & (cal_high['Sex']==3))]
cal_med_all = cal_med[((cal_med['Age']=='Total') & (cal_med['Sex']==3))]
cal_low_all = cal_low[((cal_low['Age']=='Total') & (cal_low['Sex']==3))]
```

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(14,8))
cal_all = pd.DataFrame([cal_high_all.loc[:'2020','Pop'],
                        cal_high_all.loc['2020':,'Pop'],
                        cal_med_all.loc['2020':,'Pop'],
                        cal_low_all.loc['2020':,'Pop'], 
                        ])
cal_all=cal_all.T
cal_all.columns=['Historical','High','Med','Low']
sns.lineplot(data = cal_all,ax=ax)
ax.set_title('Calgary Population Projection')
```

```python
cal_all.to_csv('cal_pop_proj.csv')
```
