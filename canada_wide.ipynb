```python
from stats_can import StatsCan

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#seaborn theme
import seaborn as sns
sns.set(font='Gill Sans MT')
```

```python
sc = StatsCan()
df = sc.table_to_df('14-10-0017-01')
#sc.update_tables()
```

```python
df
```

```python
canada = df[df['GEO']=='Canada']
```

```python
can_ft = canada[canada['Labour force characteristics']=='Full-time employment']
```

```python
emp_num = can_ft[(can_ft['Age group']=='15 years and over') & (can_ft['Sex']=='Both sexes') & (can_ft['UOM']=='Persons')]
```

```python
emp_num.set_index('REF_DATE')['2014':'2020'].plot(figsize=(14,6))
```

```python
can_ft = canada[canada['Labour force characteristics']=='Part-time employment']
emp_num = can_ft[(can_ft['Age group']=='15 years and over') & (can_ft['Sex']=='Both sexes') & (can_ft['UOM']=='Persons')]
emp_num.set_index('REF_DATE')['2014':'2020'].plot(figsize=(14,6))
```
